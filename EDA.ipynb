# Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
df = pd.read_csv('DMart.csv')
df.head()
```

# Data Cleaning

Remove null values

```python
df.isna().sum()
```

```python
df.fillna(0, inplace = True)
```

```python
df.dropna(inplace = True)
```

Drop duplicates

```python
df.duplicated().sum()
df.drop_duplicates(inplace = True)
```

# Single column analysis

Total Sales

```python
df['Price'].sum()
```

Category distribution

```python
df['Category'].value_counts()
```

Visualization

```python
top10 = df['Category'].value_counts().nlargest(10).index
plt.figure(figsize=(10,5))
sns.countplot(x='Category' , data=df , order=top10)
plt.xticks(rotation=45)
plt.show()
```

# Column VS Column

Price by column

```python
top10 = df['Category'].value_counts().nlargest(10).index
plt.figure(figsize=(13,5))
df.groupby('Category')['Price'].sum().plot(kind = 'bar')
plt.xticks(rotation = 90)
plt.show()
```
